```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Membaca data dari file CSV
df = pd.read_csv('data.csv') 

# Konversi kolom 'Gaji' menjadi tipe numerik
df['Gaji'] = df['Gaji'].replace('[\$,]', '', regex=True).astype(float)

# Melakukan filtering data berdasarkan kriteria 'Usia' di atas 30 tahun
filtered_data = df[df['Usia'] > 30]

# Fitur untuk clustering ('Gaji' dan 'Usia' sebagai contoh)
features_for_clustering = filtered_data[['Gaji', 'Usia']]  # Menggunakan data yang sudah difilter

# Inisialisasi algoritma KMeans untuk clustering
k = 3  # Jumlah klaster yang diinginkan
kmeans = KMeans(n_clusters=k)

# Melakukan clustering dengan KMeans pada data yang sudah difilter
kmeans.fit(features_for_clustering)

# Menambahkan kolom cluster ke DataFrame hasil filtering
filtered_data['Cluster'] = kmeans.labels_

# Menampilkan hasil clustering pada data yang sudah difilter
print(filtered_data[['Gaji', 'Usia', 'Cluster']])

# Visualisasi hasil clustering pada data yang sudah difilter
plt.scatter(filtered_data['Gaji'], filtered_data['Usia'], c=filtered_data['Cluster'], cmap='viridis', edgecolor='k', s=50)
plt.title('Hasil Clustering menggunakan KMeans setelah Filtering')
plt.xlabel('Gaji')
plt.ylabel('Usia')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```
